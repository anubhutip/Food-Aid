```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Importing the dataset
df = pd.read_csv('kitchen_waste.csv')
```

```python
df.head()
```

```python
df['Type of Waste'].value_counts().plot(kind='bar')
plt.xlabel('Type of Waste')
plt.ylabel('Count')
plt.title('Bar Chart of Type of Waste')
plt.show()
```

```python
df['Disposal Method'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart of Disposal Method')
plt.show()
```

```python
donated_counts = df['Donated'].value_counts()
donated_counts.plot(kind='bar', color='blue', label='Not Donated')
(df['Donated'].value_counts(normalize=True) * 100).plot(kind='bar', color='red', alpha=0.7, label='Donated %')
plt.xlabel('Donated')
plt.ylabel('Count')
plt.title('Stacked Bar Chart of Donated vs Not Donated')
plt.legend()
plt.show()
```

```python
source_of_waste = df['Source of Waste'].value_counts()
plt.pie(source_of_waste, labels=source_of_waste.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Waste by Source')
plt.show()
```

```python
waste_type = df.groupby('Type of Waste')['Weight (lbs)'].sum().reset_index()
sns.barplot(x='Type of Waste', y='Weight (lbs)', data=waste_type)
plt.title('Total Weight of Waste by Type')
plt.show()
```

```python
waste_by_month = df.groupby('Time')['Weight (lbs)'].sum().reset_index()
plt.figure(figsize=(30, 10))
sns.barplot(x='Time', y='Weight (lbs)', data=waste_by_month)
plt.title('Total Weight of Waste by Time')
plt.show()
```
