```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reading the data
df = pd.read_csv("foodwaste.csv")
```

```python
print(df.head())
```

```python
# Calculate the total amount of food waste per restaurant
total_waste_by_restaurant = df.groupby("Restaurant")["Weight (lbs)"].sum()
print(total_waste_by_restaurant)

# Create a bar plot of the total amount of food waste per restaurant
plt.figure(figsize=(30, 10))
sns.barplot(x=total_waste_by_restaurant.index, y=total_waste_by_restaurant.values)
plt.title("Total Food Waste by Restaurant")
plt.xlabel("Restaurant")
plt.ylabel("Weight (lbs)")
plt.show()
```

```python
# Calculate the amount of food waste by reason for waste
waste_by_reason = df.groupby("Reason for Waste")["Weight (lbs)"].sum()
print(waste_by_reason)

# Create a pie chart of the amount of food waste by reason for waste
plt.figure(figsize=(5, 5))
plt.pie(waste_by_reason.values, labels=waste_by_reason.index, autopct='%1.1f%%')
plt.title("Food Waste by Reason")
plt.show()
```

```python
# Calculate the amount of food waste by disposal method
waste_by_disposal = df.groupby("Disposal Method")["Weight (lbs)"].sum()
print(waste_by_disposal)

# Create a bar plot of the amount of food waste by disposal method
plt.figure(figsize=(10, 5))
sns.barplot(x=waste_by_disposal.index, y=waste_by_disposal.values)
plt.title("Food Waste by Disposal Method")
plt.xlabel("Disposal Method")
plt.ylabel("Weight (lbs)")
plt.show()
```

```python
# Calculate the total amount of food waste by food item
waste_by_food = df.groupby("Food Item")["Weight (lbs)"].sum().sort_values(ascending=False)
print(waste_by_food.head(10))

# Create a bar plot of the top 10 most wasted food items
plt.figure(figsize=(20, 5))
sns.barplot(x=waste_by_food.head(10).index, y=waste_by_food.head(10).values)
plt.title("Top 10 Most Wasted Food Items")
plt.xlabel("Food Item")
plt.ylabel("Weight (lbs)")
plt.show()
```

```python
# Calculate the total amount of food waste by restaurant and sort in descending order
waste_by_restaurant = df.groupby("Restaurant")["Weight (lbs)"].sum().sort_values(ascending=False)
print(waste_by_restaurant)

# Create a bar plot of the total amount of food waste by restaurant
plt.figure(figsize=(35, 10))
sns.barplot(x=waste_by_restaurant.index, y=waste_by_restaurant.values)
plt.title("Total Food Waste by Restaurant")
plt.xlabel("Restaurant")
plt.ylabel("Weight (lbs)")
plt.show()
```
